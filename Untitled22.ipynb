```python
%%writefile agent.py
import json
from groq import Groq
from ddgs import DDGS
import trafilatura

def search_web(query: str) -> str:
    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=5, region="in-en"))
        if not results:
            return "No results found."
        out = []
        for r in results:
            out.append(f"Title: {r.get('title')}\nURL: {r.get('href')}\nSnippet: {r.get('body')}")
        return "\n\n".join(out)
    except Exception as e:
        return f"Search error: {e}"

def extract_page(url: str) -> str:
    try:
        downloaded = trafilatura.fetch_url(url)
        if not downloaded:
            return "Could not fetch page."
        text = trafilatura.extract(downloaded)
        if not text:
            return "Could not extract content."
        return text[:3000]
    except Exception as e:
        return f"Extraction error: {e}"

TOOL_MAP = {"search_web": search_web, "extract_page": extract_page}

TOOL_SCHEMA = [
    {
        "type": "function",
        "function": {
            "name": "search_web",
            "description": "Search the web for Indian startup information",
            "parameters": {
                "type": "object",
                "properties": {"query": {"type": "string", "description": "search query"}},
                "required": ["query"],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "extract_page",
            "description": "Extract clean article text from a webpage URL",
            "parameters": {
                "type": "object",
                "properties": {"url": {"type": "string", "description": "page URL"}},
                "required": ["url"],
            },
        },
    },
]

SYSTEM_PROMPT = (
    "You are a research analyst specializing in Indian startups. "
    "Use search_web to find sources, then extract_page on the 2-3 most relevant URLs. "
    "Never invent facts — only use retrieved content. "
    "Once you have enough information, respond with ONLY the final markdown report "
    "(no tool calls) using exactly these sections:\n"
    "## Overview\n## Founders & Team\n## Funding History\n## Business Model\n"
    "## Market & Competitors\n## Recent News\n## Sources\n"
    "Under Sources, list the URLs you actually used."
)

def generate_report(startup_name: str, api_key: str, max_iterations: int = 6) -> str:
    client = Groq(api_key=api_key)
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": f"Research the Indian startup '{startup_name}': overview, founders, funding rounds, business model, competitors, recent news."},
    ]

    for _ in range(max_iterations):
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=messages,
            tools=TOOL_SCHEMA,
            tool_choice="auto",
            temperature=0.2,
        )
        msg = response.choices[0].message

        if not msg.tool_calls:
            return msg.content

        messages.append({"role": "assistant", "content": msg.content, "tool_calls": msg.tool_calls})

        for tc in msg.tool_calls:
            fn_name = tc.function.name
            args = json.loads(tc.function.arguments)
            result = TOOL_MAP[fn_name](**args)
            messages.append({
                "role": "tool",
                "tool_call_id": tc.id,
                "content": result,
            })

    return "Could not complete research within iteration limit. Try again."
```

```python
%%writefile app.py
import streamlit as st
import os
from agent import generate_report

st.set_page_config(page_title="Indian Startup Research Agent", page_icon="🚀", layout="wide")
st.title("🚀 Indian Startup Research Agent")
st.caption("Web search + extraction + structured markdown report — powered by Groq")

api_key = os.environ.get("GROQ_API_KEY") or st.text_input("Groq API Key", type="password")
startup_name = st.text_input("Startup name", placeholder="e.g. Meesho, CRED, Razorpay")

if st.button("Generate Report", type="primary"):
    if not api_key:
        st.error("Provide a Groq API key.")
    elif not startup_name:
        st.error("Enter a startup name.")
    else:
        with st.spinner("Researching... this takes ~30-60s"):
            try:
                report = generate_report(startup_name, api_key)
                st.markdown(report)
                st.download_button("Download Report (.md)", report, file_name=f"{startup_name}_report.md")
            except Exception as e:
                st.error(f"Error: {e}")
```
